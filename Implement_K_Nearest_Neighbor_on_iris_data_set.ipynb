```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
iris=datasets.load_iris()
```

```python
x=iris.data
x

```

```python
y=iris.target
y
```

```python
print(x[110],y[110])
```

```python
model=KNeighborsClassifier()
model.fit(x, y)
```

```python

result=model.predict([[6.5, 3.2, 5.1, 2.0]])
print(result)
```
